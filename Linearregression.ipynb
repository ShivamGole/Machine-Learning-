```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

df = pd.DataFrame({
    'study_hours': list(range(1, 11)),
    'exam_score': list(range(10, 101, 10))
})
print(df)
```

```python
x=df[['study_hours']]
y=df['exam_score']

# split data for train
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=42)

#  model creation
model = LinearRegression()

#  fit or train model
model.fit(xtrain,ytrain)
```

```python
#  predict the value
ypred=model.predict(xtest)
ypred
```

```python
plt.scatter(xtest,ytest)
plt.scatter(xtest,ypred)
```

```python
r2_score(ytest,ypred)
model.score(xtest,ytest)
mean_squared_error(ytest,ypred)
mean_absolute_error(ytest,ypred)
```

# Doing linear regression by pipeline approch.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# 📌 Step 1: Data
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([30000, 35000, 50000, 55000, 65000, 70000])

# 🔧 Step 2: Create pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

# 🤖 Step 3: Fit model
pipeline.fit(X, y)

# 🔮 Step 4: Predict
y_pred = pipeline.predict(X)

# 📊 Step 5: Evaluate
print(f"🔍 Coefficient (slope): {pipeline.named_steps['regressor'].coef_[0]:.2f}")
print(f"📈 Intercept: {pipeline.named_steps['regressor'].intercept_:.2f}")
print(f"📊 MSE: {mean_squared_error(y, y_pred):.2f}")
print(f"🧠 R² Score: {r2_score(y, y_pred):.2f}")

# 📉 Optional: Visualize
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted', linewidth=2)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression via Pipeline")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Data
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([30000, 35000, 50000, 55000, 65000, 70000])

# Step 2: Create pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

# Step 3: Fit model
pipeline.fit(X, y)

# Step 4: Predict
y_pred = pipeline.predict(X)

# Step 5: Evaluate
print(f"Coefficient (slope): {pipeline.named_steps['regressor'].coef_[0]:.2f}")
print(f"Intercept: {pipeline.named_steps['regressor'].intercept_:.2f}")
print(f"MSE: {mean_squared_error(y, y_pred):.2f}")
print(f"R² Score: {r2_score(y, y_pred):.2f}")

# Optional: Visualize
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted', linewidth=2)
plt.legend()
```
